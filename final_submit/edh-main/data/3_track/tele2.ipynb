```python
import pandas as pd
```

```python
tele2 = pd.read_csv('tele2_data.csv', parse_dates=['datetime'])
```

```python
tele2.shape
```

```python
tele2.head(3)
```

```python
tele2['hour'] = tele2['datetime'].apply(lambda x: x.hour)
tele2['weekday'] = tele2['datetime'].apply(lambda x: x.weekday())
tele2['year'] = tele2['datetime'].apply(lambda x: x.year)
tele2['month'] = tele2['datetime'].apply(lambda x: x.month)
```

```python
tele2.head(3)
```

```python
tele2.sort_values(by=['datetime'])
```

```python
tele2.columns
```

# Ищем количество людей в прошлом и следующем км

```python
tele2['prev_road_km'] = tele2['road_km'] -1
tele2['next_road_km'] = tele2['road_km'] +1
```

```python
pd.merge(tele2, tele2,\
         left_on = [ 'road_id', 'road_km',  'hour', 'weekday', 'year', 'month'],\
        right_on=[ 'road_id', 'prev_road_km',  'hour', 'weekday', 'year', 'month'])
```
